```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Data[1].csv')
```

```python
df
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Salary"].fillna(df["Salary"].mean(), inplace=True)

print(df)
```

```python
salary_stats = df.groupby("Country")["Salary"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)

print(salary_stats)
```

```python
salary_list = df.groupby("Country")["Salary"].apply(list)
print(salary_list)
```

```python
age_stats = df.groupby("Country")["Age"].describe()
print(age_stats)
```
